### Task_1: Restaurant Ratings

### Analyze the distribution of aggregate ratings and determine the most common rating range.

```python
df.columns
```

```python
aggregate_rating= df['Aggregate rating'].value_counts()
print(aggregate_rating)
```

```python
comman_rating= aggregate_rating.idxmax()
print('The most Comman rating range is:',comman_rating)
                                     
```

```python
plt.hist(df['Aggregate rating'],bins =30 , color = 'red',edgecolor = 'blue')
plt.xlabel('Aggregate rating')
plt.ylabel('Counts')
plt.title('Distribution Of Aggregate Ratings')
plt.show()
```

### Calculate the average number of votes received by restaurants.

```python
df.columns
```

```python
avg_votes = round(df['Votes'].mean(),2)
print('The Average Number Of Votes Received By Restaurants:',avg_votes)
```

### Task_2: Cuisine Combination

### Identify the most common combinations of cuisines in the dataset.

```python
df.columns
```

```python
common_combo = df.groupby('Cuisines')['Aggregate rating'].mean().sort_values(ascending= False)
top_20 = common_combo.head(20)
```

```python
print('Top 20 Most Combination Of Cuisines are:',top_20)
```

```python
plt.figure(figsize=(7, 6))
plt.barh(top_20['Cuisines'], top_20['Aggregate rating'], color='red')
plt.xlabel('Average Aggregate Rating')
plt.ylabel('Cuisines')
plt.title('Top 20 Cuisine With Common Combinations')
plt.gca().invert_yaxis() 
plt.show()
```

### Determine if certain cuisine combinations tend to have higher ratings..

```python
df.columns
```

```python
max_rating = common_combo.iloc[0]
print('The Max rating is:',max_rating)
```

```python
max_rated_restaurant = df.loc[df['Aggregate rating'] == max_rating]
print('Restaurant Having The Higher Rating:',max_rated_restaurant['Restaurant Name'])
```

### Task_4: Restaurant Chains
### Identify if there are any restaurant chains present in the dataset

```python
df.columns
```

```python
restaurant_chain = df.groupby('Restaurant Name').size().reset_index(name= 'Outlet_count')
new = restaurant_chain[restaurant_chain['Outlet_count']>1]
restaurant_chain = new.sort_values(by = 'Outlet_count',ascending = False)
```

```python

print(restaurant_chain[['Restaurant Name','Outlet_count']].head(15))
```

```python
plt.figure(figsize=(25, 8))
plt.bar(restaurant_chain['Restaurant Name'][:15],restaurant_chain['Outlet_count'][:15],color='red')
plt.xlabel('Restaurant Name')
plt.title('Top 15 Restaurant Chains By Number Of Outlets')
plt.ylabel('Number of outlets')
plt.show()
```

### Analyze the ratings and popularity of different restaurant chains.

```python
df.columns
```

```python
rating = df.groupby('Restaurant Name')['Aggregate rating'].mean().reset_index(name='Average rating').sort_values(by='Average rating',ascending= False)
```

```python
top_10_rating = rating.nlargest(10,'Average rating')
print("The Top 10 Restaurant Name With Average Rating:")
print(top_10_rating)
```

```python
votes = df.groupby('Restaurant Name')['Votes'].sum().reset_index(name = 'Total votes').sort_values(by= 'Total votes',ascending = False)
```

```python
top_10_votes = votes.nlargest(10,'Total votes')
print("The Top 10 Restaurant Name With Total Votes")
print(top_10_votes)
```

```python
plt.figure(figsize=(25, 4))
plt.bar(rating['Restaurant Name'][:10],rating['Average rating'][:10],color='red')
plt.xlabel('Restaurant Chain')
plt.title('Top 10 Restaurant Chains By Average Rating')
plt.ylabel('Average Rating')
plt.show()
```

```python
plt.figure(figsize=(21, 5))
plt.bar(votes['Restaurant Name'][:10],votes['Total votes'][:10],color='red')
plt.xlabel('Restaurant Chain')
plt.title('Top 10 Restaurant Chains By Total Votes')
plt.ylabel('Total Votes')
plt.show()
```
